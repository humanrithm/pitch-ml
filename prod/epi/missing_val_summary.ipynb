```python
import pickle
import pandas as pd
```

$\textbf{Epidemiology: Missing Value Imputation Summary}$

```python
# load imputer summary
with open('storage/imputer_summary.pkl', 'rb') as f:
    imputer_summary = pickle.load(f)
```

```python
means_df = pd.DataFrame({
    'nonimputed_means': imputer_summary['nonimputed_means'][2015],
    'imputed_means': imputer_summary['imputed_means'][2015]
})
means_df
```

```python
# compute and store diffs btw imputed and nonimputed means
all_diffs = []
for y in imputer_summary['nonimputed_means'].keys():
    diffs = imputer_summary['imputed_means'][y] - imputer_summary['nonimputed_means'][y]
    
    # append to list
    all_diffs.append(diffs)

# create dframe
diffs_df = pd.DataFrame(all_diffs, index=imputer_summary['nonimputed_means'].keys()).reset_index(names='year')
```
